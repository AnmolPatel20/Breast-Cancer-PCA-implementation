## Breast Cancer Prediction using PCA(Principal Component Analysis)
#### Breast cancer wisconsin (diagnostic) dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
#loading dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset = load_breast_cancer()
```

```python
print(type(cancer_dataset))
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
df=pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
df.head()
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
scaled_data
```

```python
##  Applying PCA Algorithms
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
data_pca=pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_dataset['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
plt.colorbar(label='0 = Malignant, 1 = Benign')
plt.title('PCA of Breast Cancer Dataset')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(data_pca, cancer_dataset['target'], test_size=0.2, random_state=42)
model = LogisticRegression()
model.fit(X_train, y_train)
print(classification_report(y_test, model.predict(X_test)))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, display_labels=['Malignant', 'Benign'])
plt.title('Confusion Matrix')
plt.show()
```

```python
# Take one sample from the original dataset
new_sample = scaled_data[0].reshape(1, -1)

# Transform it through PCA first
new_sample_pca = pca.transform(new_sample)
```

```python
# Predict
prediction = model.predict(new_sample_pca)
print("Malignant" if prediction[0] == 0 else "Benign")
```
